# Set up

```python
import csv
import numpy as np
import pandas as pd
import seaborn as sns # for visualiation
from scipy.stats import ttest_ind # t-tests
import statsmodels.formula.api as smf # linear modeling
import statsmodels.api as sm
import matplotlib.pyplot as plt # plotting
import matplotlib
matplotlib.style.use('ggplot')
%matplotlib inline 
```

```python
# Read data
df = pd.read_csv('./data/Youth_Tobacco_Survey__YTS__Data.csv')
```

# Data Preparation

I dropped some unnecessary variables in this section. For example, those variables are all the same for all rows. I also dropped some rows that do not have the data value column, because data in these cells have been suppressed because of a small sample size and I think it is not necessary to include them. 

```python
df = df.drop('TopicType', 1)
df = df.drop('DataSource', 1) 
df = df.drop('Data_Value_Unit', 1)
df = df.drop('Data_Value_Type', 1)
df = df.drop('Race', 1)
df = df.drop('Age', 1)   
df = df.drop('TopicTypeId', 1)
df = df.drop('StratificationID2', 1)
df = df.drop('StratificationID3', 1)
df = df[np.isfinite(df['Data_Value'])]
df
```

# EDA: Describing Data Structure

```python
cn = len(df.columns)
rn = len(df.index)
print(cn, rn)

list(df.columns)
```

# EDA: Univariate Analysis

The first graph is about the general smoking rate among all the survey date. I can tell that smoking rate below 10 percent has a great numbers of people. 

```python
plt.hist(df.Data_Value)
plt.title('Smoking rate')
plt.ylabel('numbers of data')
plt.xlabel('percentage')
plt.show()
```

```python
df['Data_Value'].mean()
```

```python
df['Data_Value'].median()
```

```python
df['Data_Value'].std()
```

```python
ma = df['Data_Value'].max()
mi = df['Data_Value'].min()
ran = ma - mi
ran
```

In this graph, I found that the data were obtained during countinuous years while in the earliest years with the most data obtained. 

```python
plt.hist(df.YEAR)
plt.title('Years of when data was obtained')
plt.ylabel('numbers of data')
plt.xlabel('year')
plt.show()
```

In this graph, I found that the low confidence limit and high confidence limit do not have a huge gap between most of them. They all follow the general trend of the smoking rate.

```python
plt.hist(df.Low_Confidence_Limit, alpha=0.5, label='Low Confidence Limit')
plt.hist(df.High_Confidence_Limit, alpha=0.5, label='High Confidence Limit')
plt.legend(loc='upper right')
```

# EDA: Univariate Analysis by Category

In this graph, I put all the smoking rate for all genders together in order to see the differences. And I found the differences between female and male. 

```python
x=df['Data_Value'][df['Gender'] == 'Male']
y=df['Data_Value'][df['Gender'] == 'Female']
# z=df['Data_Value'][df['Gender'] == 'Overall']

plt.hist(x, alpha=0.5, label='male')
plt.hist(y, alpha=0.5, label='female')
# plt.hist(z, alpha=0.5, label='overall')
plt.legend(loc='upper right')
```

In this graph, I put all the smoking rate for all education level together in order to see the differences. And I found the differences between high school youths and middle school youths.

```python
x=df['Data_Value'][df['Education'] == 'Middle School']
y=df['Data_Value'][df['Education'] == 'High School']

plt.hist(x, alpha=0.5, label='middle school')
plt.hist(y, alpha=0.5, label='high school')
plt.legend(loc='upper right')
```

In this graph, I put all the smoking rate for all different topics together in order to see the differences. And I found the differences between those survey topics.

```python
x=df['Data_Value'][df['TopicDesc'] == 'Cessation (Youth)']
y=df['Data_Value'][df['TopicDesc'] == 'Cigarette Use (Youth)']
z=df['Data_Value'][df['TopicDesc'] == 'Smokeless Tobacco Use (Youth)']

plt.hist(x, alpha=0.5, label='Cessation (Youth)')
plt.hist(y, alpha=0.5, label='Cigarette Use (Youth)')
plt.hist(z, alpha=0.5, label='Smokeless Tobacco Use (Youth)')
plt.legend(loc='upper right')
```
