## EDA: Univariate Analysis by Category

In this section, you should investigate how each your dependent variable of interest varies across groupings by your independent variable(s).
```python
df = df[df['TopicDesc'].isin(['Cessation (Youth)'])]

df_quit = df[df['MeasureDesc'] == 'Percent of Current Smokers Who Want to Quit']
df_attempt_quit = df[df['MeasureDesc'] == 'Quit Attempt in Past Year Among Current Cigarette Smokers']
```

The main thing this report will analyze is cessation.  
First, seperate the group to Current Smokers Who Want to Quit and Quit Attempt in Past Year Among Current Cigarette Smokers.  
Let's look at middle school education leve and all gender. We evaluate the trend of the mean of all states over years on people who want to quit.

```python
#Education level: Middle School 
#Gender:overall 
#Mean of all states
#Trends of people want to quit over years 

df_overall_middle = df_quit[(df_quit['Gender'] == 'Overall') & (df_quit['Education'] == 'Middle School')]
a = df_overall_middle.groupby('YEAR').mean().reset_index()

year = a.YEAR
mean = a.Data_Value

matplotlib.pyplot.scatter(year,mean)

matplotlib.pyplot.show()
```

Then we evaluate the trend of the mean of all states over years on people who attempt to quit before.

```python
#Trends of people attempt to quit over years 

df_overall_middle_a = df_attempt_quit[(df_attempt_quit['Gender'] == 'Overall') & (df_attempt_quit['Education'] == 'Middle School')]
b = df_overall_middle_a.groupby('YEAR').mean().reset_index()

year = b.YEAR
mean = b.Data_Value

matplotlib.pyplot.scatter(year,mean)

matplotlib.pyplot.show()
```

We will evaluate the same thing for high school education level. 

```python
#Education level: High School 
#Gender:overall 
#Mean of all states
#Trends of people want to quit over years 


df_overall_high = df_quit[(df_quit['Gender'] == 'Overall') & (df_quit['Education'] == 'High School')]
c = df_overall_high.groupby('YEAR').mean().reset_index()
year = c.YEAR
mean = c.Data_Value

matplotlib.pyplot.scatter(year,mean)

matplotlib.pyplot.show()
```

```python
#Trends of people attempt to quit over years 

df_overall_high_a = df_attempt_quit[(df_attempt_quit['Gender'] == 'Overall') & (df_attempt_quit['Education'] == 'High School')]
d = df_overall_high_a.groupby('YEAR').mean().reset_index()
year = d.YEAR
mean = d.Data_Value

matplotlib.pyplot.scatter(year,mean)

matplotlib.pyplot.show()
```

## EDA: Univariate Analysis


In this step, we will evaluate for every states, their percentage trends of people who want to quit and people who attempt to quit. We seperate them by educational level as well. We divide the data to people who want to quit and who attempted to quit before, and then in each grooup further divided them by middle school education level and high school education level.

```python
#every states 
#Trend over years
#Education:Middle School 
#Gender:Overall 
#Want to quit

df_overall_middle = df_quit[(df_quit['Gender'] == 'Overall') & (df_quit['Education'] == 'Middle School')]

location=df_overall_middle.sort_values(by = ['LocationAbbr','YEAR'])
location_year = location[['LocationAbbr','YEAR','Data_Value']]
```

```python
from ggplot import *
x = ggplot(location_year, aes(x="YEAR", y="Data_Value"))
x + geom_point() + \
    facet_wrap("LocationAbbr")
```
