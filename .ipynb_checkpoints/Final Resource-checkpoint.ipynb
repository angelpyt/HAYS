# Youth Tobacco Survey Analysis
In this resource, we provide supporting data, graphs, and evidence of important trends in youth tobacco usage to get attention from parents, schools, organizations, and the government.

The dataset we are using is the Youth Tobacco Survey (YTS) Data, which has been conducted by the Centers for Disease Control and Prevention (CDC) yearly since 1998. The dataset we accessed through Data.gov, where CDC has published and made it publicly accessible, includes only data for years between 1999 to 2015.

This resource provides information on:
- The overall trend of youth tobacco usage from 1999 to 2015.
- How the youth smoking users and trends differ in different states in the U.S.
- How the smoking status differs between males and females.
- How the percentage of current smokers who want to quit varies throughout the years



## Current Smoker Analysis
In this section we analyze current smoker trends for youth between 1999-2015. We focus on gender and education as two primary factors in our analysis. We also look at geographic location but decided that it isn't an important factor to continue with moving forward.
### Current Youth Smoker Percentage by Year and Gender
First we graph the mean overall percentage of current smokers. All states and education are combined and it is split between Male and Female self-identified gender. Here we can clearly see the overall decreasing trend over time for both males and females. 

```python
# smoking status data by states
smokingstat_df = df[df['MeasureDesc'] == "Smoking Status"]
current_smoker_df = smokingstat_df[smokingstat_df['Response'] == 'Current']
# current_smoker_df
```

```python
# Mean Data_Value by sex & year
mean = current_smoker_df[['Gender', 'Data_Value', 'YEAR']].groupby(['Gender','YEAR']).agg('mean').reset_index()
female = current_smoker_df[['Gender', 'Data_Value', 'YEAR']].groupby(['Gender']).get_group('Female')
female_current_smoker_mean_by_year = female.groupby('YEAR').agg('mean').reset_index()
male = current_smoker_df[['Gender', 'Data_Value', 'YEAR']].groupby(['Gender']).get_group('Male')
male_current_smoker_mean_by_year = male.groupby('YEAR').agg('mean').reset_index()
```

```python
trace_current_male_smoker = go.Scatter(
    x = male_current_smoker_mean_by_year.YEAR,
    y = male_current_smoker_mean_by_year.Data_Value,
    name = 'male',
)

trace_current_female_smoker = go.Scatter(
    x = female_current_smoker_mean_by_year.YEAR,
    y = female_current_smoker_mean_by_year.Data_Value,
    name = 'female',
    line = dict(
        color = ('rgb(205, 12, 24)'),)
)

# Edit the layout
layout = dict(title = 'Percentage of current youth tobacco user (by gender)',
              xaxis = dict(title = 'Year'),
              yaxis = dict(title = 'Percentage'),
              )
```

```python
data = [trace_current_male_smoker, trace_current_female_smoker]

fig = dict(data=data, layout=layout)
py.iplot(fig, filename = 'Percentage of current youth tobacco user (by gender)')
```

### Linear model Fitting Current Smokers Throughout the Past Decade
The adjusted R-squared for current smokers is 0.759, indicating good fit of the model to the data. Our P-values are also significant between current smoking rate and males and between current smoking rate and middle school. The coefficient for males is positive while it is negative for middle school education. This indicates that there is a significant positive relationship between males and currently smoking status, and a significant negative relationship between middle schoolers and current smoking.

```python
current_smoker_df['Data_Value'] = current_smoker_df['Data_Value'] / 100
```

```python
lm = smf.ols(formula='Data_Value ~ YEAR + Gender + Education', data=current_smoker_df).fit()
lm.summary()
```

```python
# Create predictions using this model
current_smoker_df['mult_preds'] = lm.predict()
```

Then we test the model fit with the actual values by plotting the predicted percentage on the x-axis with the actual recorded percentage on the y-axis. Therefore, the  see how the predictions matching the actual data. x-axis is the predicted value, and the y-axis is the actual value. 

```python
# Compare predictions to observations
plt.scatter(current_smoker_df.mult_preds, current_smoker_df.Data_Value, color = 'blue', alpha = 0.5)
plt.plot(current_smoker_df.Data_Value, current_smoker_df.Data_Value)
plt.xlabel("Predicted Percentage")
plt.ylabel("Actual Percentage")
plt.show()
```

```python
# Visually compare predictions over year
plt.scatter(current_smoker_df.YEAR, current_smoker_df.Data_Value, color = 'blue')
plt.scatter(current_smoker_df.YEAR, current_smoker_df.mult_preds, color='red')
plt.show()
```

### Linear model fitting smokers attempt to quit

```python
df_quit['Data_Value'] = df_quit['Data_Value'] / 100
```

```python
lm_quit = smf.ols(formula='Data_Value ~ YEAR + Gender + Education', data=df_quit).fit()
lm_quit.summary()
```

```python
# Create predictions using this model
df_quit['mult_preds'] = lm_quit.predict()
```

```python
# Compare predictions to observations
plt.scatter(df_quit.YEAR, df_quit.Data_Value, color = 'blue', alpha = 0.5, label='Actual')
plt.scatter(df_quit.YEAR, df_quit.mult_preds, color = 'red', alpha = 0.5, label='Predicted')
#plt.ylim((0, 0.1))
plt.xlabel("Year")
plt.ylabel("Percentage of Smokers who Attempt Quit")
plt.legend(loc='upper right')
plt.show()
```
